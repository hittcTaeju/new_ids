```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('preprocessing_cicids2017.csv')
```

```python
df
```

```python
# 상수값 컬럼 제거
df = df.loc[:, (df != df.iloc[0]).any()]
df
```

```python
corr_matrix = df.drop('Attack Type', axis=1).corr()
corr_matrix.style.background_gradient(cmap = 'coolwarm', axis = None).format(precision = 2)
```

```python
# 상관관계 임계값 설정 (예: 0.9 이상은 상관관계가 높다고 판단)
threshold = 0.9

# 상관관계가 높은 피처 목록 추출
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# 상관계수가 threshold 이상인 컬럼 추출
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

# 상관관계 높은 컬럼을 제거한 데이터프레임 생성
df_reduced = df.drop(columns=to_drop)

# 결과 확인
print(f"제거된 컬럼: {to_drop}")
print(df_reduced.head())
```
